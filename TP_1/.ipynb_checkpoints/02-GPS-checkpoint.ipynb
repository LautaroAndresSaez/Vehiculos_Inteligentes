```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import plotly.graph_objects as go
from matplotlib import animation
```

```python
from os.path import join
folder_path = './dataset/gps'
data_time = ['14_33_59.txt',
             '14_40_43.txt',
             '14_43_27.txt',
             '14_45_40.txt',
             '14_57_29.txt',
             '14_59_21.txt',
             '15_05_19.txt',
             '15_12_18.txt',
             '15_15_23.txt',
             '15_27_43.txt',
             '15_32_31.txt',
             '15_36_58.txt',
            ]
static_pref = './GPS_2013_12_12_'
```

```python
gps_dfs = []
for data in data_time:
    df = pd.read_csv( join( folder_path, static_pref + data ), sep=',' )
    del df['Message ID']
    del df['Diff Ref Station ID']
    gps_dfs.append( df )
len( gps_dfs )
```
